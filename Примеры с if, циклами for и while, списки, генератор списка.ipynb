## Задача 1. Цикл for, условный оператор if, списки
**Входные данные**
В первой строке входного файла INPUT.TXT записано натуральное число N (1 ≤ N ≤ 100) – число монеток. В каждой из последующих N строк содержится одно целое число – 1 если монетка лежит решкой вверх и 0 если вверх гербом.

**Выходные данные**
В выходной файл OUTPUT.TXT выведите минимальное количество монет, которые нужно перевернуть.

```python
n = int(input())
lst = []
for i in range (0,n):
    m = int(input()) 
    lst.append(m) #добавление элементов в список

money_zero = 0
money_one = 0
for i in lst:
    if i == 0:
        money_zero += 1
    else:
        money_one += 1
if money_zero < money_one:      
    print('Количество монеток, которое нужно перевернуть:', money_zero)
else:
    print('Количество монеток, которое нужно перевернуть:', money_one)
```

## Задача 2. Цикл for, условный оператор if, списки, операторы break, continue

Напишите программу, реализующую алгоритм перевода целого положительного числа NN из двоичной системы счисления в десятичную.

**На вход программе** подаётся число NN, представленное в двоичной системе счисления.

**Программа должна вывести** число NN, представленное в десятичной системе счисления.

```python
num = input()
num = num[::-1] #строка в обратном порядке
n = len(num) #подсчет символов в строке
k = 0 #переменная индекса и степени
s = 0 #переменная для записи десятичных значений
lst = [] #список, чтобы посчитать десятичные значения потом
for i in range(n):
    if num[k] == '1': #важно понимать, что мы работаем с данными типа СТРОКА, поэтому кавычки
        s = 2**k
        lst.append(s) 
        k+=1
    else:
        k+=1
        continue
lst = [int(i) for i in lst] #генератор списка
summ = sum(lst)
print(summ)
```

## Задача 3. Применение цикла while, логические операторы 

Напишите программу, которая считывает целые числа с консоли по одному числу в строке.

Для каждого введённого числа проверить:
если число меньше 10, то пропускаем это число;
если число больше 100, то прекращаем считывать числа;
в остальных случаях вывести это число обратно на консоль в отдельной строке.

```python
while True:
    a=int(input())
    if a<10:
        continue
    elif a>100:
        break
    else:
        print (a)
```

## Задача 4. Применение цикла while, генератор списка

Напишите программу, которая **считывает с консоли числа** (по одному в строке) до тех пор, **пока сумма введённых чисел не будет равна 0** и сразу после этого выводит сумму квадратов всех считанных чисел.

Гарантируется, что в какой-то момент сумма введённых чисел окажется равной 0, после этого считывание продолжать не нужно.

В примере мы считываем числа 1, -3, 5, -6, -10, 13; в этот момент замечаем, что сумма этих чисел равна нулю и выводим сумму их квадратов, не обращая внимания на то, что остались ещё не прочитанные значения.

```python
a = [int (i) for i in input().split()] #пользователь вводит значение, оно заносится в список а
s = sum(a) #ввели переменную, в которой будут накапливаться введенные пользователем значения
while s!=0: #пока накопитель не равен нулю, выполняется следующий цикл
    a.append(int(input())) #пользователь добавляет значение в список а
    s=sum(a) #сумматор суммирует значения
x = [i*i for i in (a)] #список квадратов значений
Xn = sum (x) #сумматор квадратов значений
print (Xn)
```
